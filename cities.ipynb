```python
import pandas as pd
import matplotlib.pyplot as plt

# Read CSV
df = pd.read_csv('cities.csv')

# Rename column if needed
df.rename(columns={'Population (Est.)': 'Population'}, inplace=True)

# Remove commas and convert to integer
df['Population'] = df['Population'].str.replace(',', '').astype(int)

# Select top 5 by population
top5 = df.sort_values(by='Population', ascending=False).head(5)

# Sort ascending for plotting from smallest to largest
top5 = top5.sort_values(by='Population', ascending=True)

# Plot
plt.figure(figsize=(10, 6))
plt.bar(top5['City'], top5['Population'], color='lightgreen')

plt.xticks(rotation=45)
plt.title('Population of Top 5 Cities')
plt.xlabel('City')
plt.ylabel('Population')

# Format y-axis with commas
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{int(x):,}'))

plt.tight_layout()
plt.show()

```
